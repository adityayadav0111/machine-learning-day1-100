```python
import numpy as np
import pandas as pd
from sklearn.experimental import enable_iterative_imputer  # noqa
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# The comment # noqa is used to tell inters (tools that analyze code for errors or style issues) 
# to ignore the line it’s attached to. "No QA" stands for "No Quality Assurance," meaning "don’t check this line."
```

```python
df = pd.read_csv(r"C:\Users\Admin\Documents\Jupyter Notebook\Machine Learning (CampusX)\Dataset\50_Startups.csv",na_values=[0])
```

```python
df.drop('State',axis=1,inplace=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop('Profit',axis=1),df['Profit'],test_size=0.3)
```

```python
imputer = IterativeImputer(estimator=LinearRegression(), max_iter=10, random_state=0)
```

```python
X_train = pd.DataFrame(imputer.fit_transform(X_train), columns=X_train.columns)
```

```python
X_train
```

```python
X_test = pd.DataFrame(imputer.transform(X_test), columns=X_train.columns)
```

```python
lr = LinearRegression()

lr.fit(X_train,y_train)

y_pred2 = lr.predict(X_test)

r2_score(y_test,y_pred2)*100
```
