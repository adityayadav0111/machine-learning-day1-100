## Power Transformer

## Types
- **Box- Cox** : Box and cox are person who introduced this transformation
1. xi ** lambda = Xi**lambda -1/lambda if lambda is not 0 if lambda =0
2. The exponent here is lambda that varies over the range -5 to 5(highest-lowest), and in the process of searching, we examine all values of lambda. finally, we choose the optimal value (resulting in best approximation to a normal distribution) for your varible.
3. Maximum liklyhood, bayesian stats.
4. strictly applicable to x>0 and no negative numbers
5. Linear model ke liye must
   
- **Yeo-Johnson**
1. It is applicable to 0s and negative numbers
2. This transformation is somewhat of an adjustment to the Box-Cox transformation, by which we can apply it to negative numbers.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PowerTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv(r"C:\Users\Admin\Documents\Jupyter Notebook\Machine Learning (CampusX)\Dataset\concrete_data.csv")
```

```python
X= df.drop('Strength',axis=1)
y = df['Strength']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=128)
```

```python
lr = LinearRegression()

lr.fit(X_train,y_train)

y_pred = lr.predict(X_test)

print("r2 score is :",r2_score(y_test,y_pred))
```

```python
## Cross Val score
lr2 = LinearRegression()
np.mean(cross_val_score(lr2,X,y,scoring='r2'))
```

```python
## Plotting the distplot without transformation

for col in X_train.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(X_train[col])
    plt.title(col)
    plt.subplot(122)
    stats.probplot(X_train[col],dist='norm',plot=plt)
    plt.title(col)
    plt.show()
```

```python
## Applying Box Cox   by default method is yeo johnson

pt = PowerTransformer(method='box-cox')

X_train_transform = pt.fit_transform(X_train+0.000001)
X_test_transform = pt.transform(X_test+0.000001)

pd.DataFrame({'cols':X_train.columns,'box_cox_lambdas':pt.lambdas_})
```

```python
### Applying Linear Regression on transformed data
lr2 = LinearRegression()

lr.fit(X_train_transform,y_train)

y_pred = lr.predict(X_test_transform)

print("R2 score is :",r2_score(y_test,y_pred))
```

```python
# using cross val score
pt = PowerTransformer(method='box-cox')
X_transformed = pt.fit_transform(X+0.0001)

lr3 = LinearRegression()
np.mean(cross_val_score(lr3,X_transformed,y,scoring='r2'))
```

```python
# Before and  after comparison for box cox
X_train_transform = pd.DataFrame(X_train_transform,columns=X_train.columns)  ##Since this is standardized and converted to nnumpy array we need to convert back to df


for i in X_train_transform.columns:
    plt.figure(figsize=(12,4))
    plt.subplot(121)
    sns.distplot(x=X_train[i])
    plt.title("Box cox dist before")
    

    plt.subplot(122)
    sns.distplot(x=X_train_transform[i])
    plt.title("box cox after transformation")
    plt.show()
```
