```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
import numpy as np
```

```python
X,y= fetch_california_housing(return_X_y=True)
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
lr2= LinearRegression()
```

```python
estimators = [("lr",lr),("dt",dt),("lr2",lr2)]
```

```python
for i in estimators:
  x = cross_val_score(i[1],X,y,cv=10,scoring="r2")
  print(i[0],np.round(np.mean(x),2))
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vc = VotingRegressor(estimators=estimators)
x = cross_val_score(vc,X,y,cv=10,scoring="r2")
print(np.round(np.mean(x),2))
```

```python
## Applying weights
for i in range(1,4):
  for j in range(1,4):
    for k in range(1,4):
      vc = VotingRegressor(estimators=estimators,weights=[i,j,k])
      x = cross_val_score(vc,X,y,cv=10,scoring="r2")
      print("for i ={}, j={}, k= {}".format(i,j,k),np.round(np.mean(x),2))
```

```python
## Using same algorithm
dt1 = DecisionTreeRegressor(max_depth=1)
dt2 = DecisionTreeRegressor(max_depth=3)
dt3 = DecisionTreeRegressor(max_depth=5)
dt4 = DecisionTreeRegressor(max_depth=7)
dt5 = DecisionTreeRegressor(max_depth=None)
```

```python
estimators = [('dt1',dt1),('dt2',dt2),('dt3',dt3),('dt4',dt4),('dt5',dt5)]
```

```python
for estimator in estimators:
  scores = cross_val_score(estimator[1],X,y,scoring='r2',cv=10)
  print(estimator[0],np.round(np.mean(scores),2))
```

```python
vr = VotingRegressor(estimators)
scores = cross_val_score(vr,X,y,scoring='r2',cv=10)
print("Voting Regressor",np.round(np.mean(scores),2))
```
