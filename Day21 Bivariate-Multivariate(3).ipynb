## 2. Bar Plot (Categorical-Numerical)

```python
df = pd.read_csv(r'C:\Users\Admin\Documents\Jupyter Notebook\Machine Learning (CampusX)\Dataset\train.csv')
```

```python
sns.barplot(x='Pclass',y='Age',data=df)   ##Average Age Vs Pclass    ##Confidence interval 
```

What Happens in the Plot
Grouping: Seaborn first groups the data based on the unique values in Pclass (the x-axis). In this case, the groups will be:

Group 1: All passengers in 1st class.
Group 2: All passengers in 2nd class.
Group 3: All passengers in 3rd class.
Averaging: For each group, Seaborn calculates the average age of passengers. It doesn't plot individual ages; instead, it takes all the ages within each class, adds them together, and divides by the number of passengers in that class to get the average.

Bar Heights: The height of each bar represents this average age for the corresponding class.

For example, if the average age of passengers in 1st class is 35 years, the bar for 1st class will go up to 35 on the y-axis.
Confidence Intervals: By default, Seaborn also shows confidence intervals on top of the bars as black lines. These lines give you an idea of the variability in the datahow much the ages of passengers in each class might vary around the average.

## 3. Boxplot (Categorical-Numerical)

```python
sns.boxplot(x='Sex',y='Age',data=df,hue='Survived')
```

## 4. Distplot (Categorical-Numerical)

```python
sns.distplot(x=df[df['Survived']==1]['Age'],hist=False)
sns.distplot(x=df[df['Survived']==0]['Age'],color='r',hist=False)
```

## 5. Heatmap (Categorical Categorical)

```python
sns.heatmap(pd.crosstab(df['Pclass'],df['Survived']))
```

## 6. Clustermap (CATEGORICAL - CATEGORICAL)

```python
sns.clustermap(pd.crosstab(df['Pclass'],df['Survived']))   #dendogram
```
