## 2. Text Cleaning
1. Remove HTML tags
2. Remove Special Characters
3. Converting every thing to lower case
4. Remove stop words
5. stemming

```python
## 1. Removing HTML tags
import re
clean = re.compile('<.*?>')
re.sub(clean,"",df.iloc[1].review)
```

```python
## function to clean html tags
def remove_html(data):
  clean = re.compile('<.*?>')
  return re.sub(clean,'',data)
```

```python
df['review'] = df['review'].apply(remove_html)
```

```python
## converting text into lower case
def lower(data):
  return data.lower()
```

```python
df['review'] = df['review'].apply(lower)
```

```python
df['review'][1]
```

```python
## to remove special character
text = "I am a $ant"

def remove_special(text):
  x = ""
  for i in text:
    if i.isalnum():
      x = x+i
    else:
      x=x+" "

  print(x)
```

```python
remove_special(df['review'])
```

```python
df['review'][1]
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
print(stop_words)
```

```python
## removing stopwords
def remove_srwds(data):

  x= []
  for i in data.split():
    if i not in stop_words:
      x.append(i)
  y = x[:]
  x.clear()
  return y
```

```python
df['review'] = df['review'].apply(remove_srwds)
```

```python
df['review']
```

```python
## applying stemming
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
y = []

def stemmer(data):
  for i in data:
    y.append(ps.stem(i))

  z = y[:]
  y.clear()
  return z
```

```python
df['review'] = df['review'].apply(stemmer)
```

```python
stemmer(['I','Love','loving','lover','loved','loves'])
```

```python
## joining back text
def join_back(data):
  return " ".join(data)
```

```python
df['review'] = df['review'].apply(join_back)
```

```python
df.review
```

```python
## most frequent words in corpus
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000)
```

```python
X = cv.fit_transform(df['review']).toarray()
```

## 4. Train Test Split

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.3)
```

## 5. Modelling

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

```python
gnb.fit(X_train,y_train)
mnb.fit(X_train,y_train)
bnb.fit(X_train,y_train)
```

```python
y_pred1 = gnb.predict(X_test)
y_pred2 = mnb.predict(X_test)
y_pred3 = bnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred1))
print(accuracy_score(y_test,y_pred2))
print(accuracy_score(y_test,y_pred3))
```
