## Normalization
- Normailzation is a technique often applied as part of data preparation for ML. The goal of normalization is to change the values of numeric columns in the dataset to use a common scale, without distorting difference in the ranges of values or losing information
- Units for weight should be removed here
- Rescales data to a range, typically between 0 and 1.

Let's simplify this. 

### Normalization (Min-Max Scaling)
Think of normalization as **squeezing data into a specific range**, usually between 0 and 1. 

- **When to Use**:
  - If you want all data values to fall between 0 and 1, which helps prevent certain values from dominating others in terms of scale.
  - For models that work based on **distances** (like k-Nearest Neighbors and Neural Networks). For these models, having all features in a similar range ensures no one feature overpowers others just because it has larger numbers.

- **Example**:
  Say you have a dataset where `Age` ranges from 20 to 70 and `Income` ranges from $30,000 to $100,000. Without scaling, `Income` would have a larger influence just because the values are bigger. With normalization, both `Age` and `Income` get scaled between 0 and 1, balancing their influence.

  ```python
  from sklearn.preprocessing import MinMaxScaler

  scaler = MinMaxScaler()
  normalized_data = scaler.fit_transform(data)  # Now all values lie between 0 and 1
  ```

---

### Standardization (Z-Score Scaling)
Standardization, on the other hand, **centers the data around 0 with a spread of 1**.

- **When to Use**:
  - If your data contains **outliers** or you expect it to be normally distributed (bell-shaped), standardization often works better.
  - For models like **linear regression, logistic regression, and PCA**, which assume the data is centered around zero.

- **Example**:
  With standardization, if you have a dataset with `Height` and `Weight`, both of these features will be scaled to have a mean of 0 and a standard deviation of 1. This makes features comparable without forcing them into a specific range.

  ```python
  from sklearn.preprocessing import StandardScaler

  scaler = StandardScaler()
  standardized_data = scaler.fit_transform(data)  # Mean = 0, Std Dev = 1
  ```

---

### Which to Choose?
- **Normalization** is useful for neural networks and distance-based algorithms.
- **Standardization** is useful for linear models and algorithms assuming normally distributed data.
